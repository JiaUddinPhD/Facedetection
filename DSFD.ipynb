```
import cv2
import face_detection
import matplotlib.pyplot as plt
print(face_detection.available_detectors)
detector = face_detection.build_detector(
  "DSFDDetector", confidence_threshold=.5, nms_iou_threshold=.3)
# BGR to RGB

import os 
source_list=os.listdir('/content/drive/MyDrive/raw_data_2_output/cluster0/Orginal-c0 ')
for f in source_list:
  f_name=os.path.join('/content/drive/MyDrive/raw_data_2_output/cluster0/Orginal-c0 ', f)
  im = cv2.imread(f_name)[:, :, ::-1]
  print(f)
  detections = detector.detect(im)
  for row_index in range(detections.shape[0]):
        ann = detections[row_index]
        x1 = int(ann[0])
        y1 = int(ann[1])
        x2 = int(ann[2])
        y2 = int(ann[3])
        
        cropped_face = im[y1:y2, x1:x2]
        
        if cropped_face.shape[0] == 0 or cropped_face.shape[1] == 0:
          print(f"Detector extracted a face with 0 rows or columns, skipping this one for {fn}")
          continue
        else:
          #print(f_name)
          #print(cropped_face)
          plt.axis("off")
          imgplot = plt.imshow(cropped_face)
          plt.savefig('/content/drive/MyDrive/raw_data_2_output/cluster0/results/'+f)  
          #plt.savefig(path + str(np.argmax(U_L[i],axis=0)) + '/' + str(i) + )
```
